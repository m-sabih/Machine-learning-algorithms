# K- Mean Clustering

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as img
import math
from PIL import Image

```

### Loading Image 1

```python
image1 = np.array(Image.open('11.jpg'))
print(image1.shape)

rows, col, channel = image1.shape
plt.imshow(image1);
```

```python
image1_reshaped=image1.reshape(-1,3)
print(image1_reshaped.shape)
```

```python
print(image1_reshaped)
```

## For k=2

```python
p1=np.random.randint(0,255,(1,3))
p2=np.random.randint(0,255,(1,3))
p1_cluster=p2_cluster=np.empty((1,3))
p1_old=p2_old=np.zeros((1,3))
image1_with_clusters=np.empty((image1_reshaped.shape[0],1))
epoch=0
flag="true"
print(p1)
print(p2)
image1_k=np.copy(image1_reshaped)
while flag=="true":
    i=0
    if np.allclose(p1_old,p1,rtol=0.01,atol=0,equal_nan=True) and np.allclose(p2_old,p2,rtol=0.01,atol=0,equal_nan=True):
        flag="false"
    else:
        p1_old,p2_old=p1,p2
        epoch+=1
        print("\nepoch ",epoch,":\n")
        for row in image1_reshaped:
            diff_p1=abs(np.sum(row-p1))
            diff_p2=abs(np.sum(row-p2))
            #print(diff_p1, diff_p2)
            if diff_p1<diff_p2:
                image1_with_clusters[i]=1
                p1_cluster=np.vstack((p1_cluster,row))
            else:
                image1_with_clusters[i]=2
                p2_cluster=np.vstack((p2_cluster,row))
            i+=1
        p1=np.mean(p1_cluster, axis=0)
        p2=np.mean(p2_cluster, axis=0)
        print(p1)
        print(p2)
```

```python
for j in range(image1_reshaped.shape[0]):
    if image1_with_clusters[j]==1.0:
        image1_k[j]=p1
    elif image1_with_clusters[j]==2.0:
        image1_k[j]=p2
```

```python
image1_k=image1_k.reshape(image1.shape)
plt.imshow(image1_k)
plt.show()
```

## For k=5

```python
p1=np.random.randint(0,255,(1,3))
p2=np.random.randint(0,255,(1,3))
p3=np.random.randint(0,255,(1,3))
p4=np.random.randint(0,255,(1,3))
p5=np.random.randint(0,255,(1,3))
p1_cluster=p2_cluster=p3_cluster=p4_cluster=p5_cluster=np.empty((1,3))
p1_old=p2_old=p3_old=p4_old=p5_old=np.zeros((1,3))
image1_with_clusters=np.empty((image1_reshaped.shape[0],1))

epoch=0
flag="true"
print(p1)
print(p2)
print(p3)
print(p4)
print(p5)

image1_k=np.copy(image1_reshaped)
while flag=="true":
    if np.allclose(p1_old,p1,rtol=0.01,atol=0,equal_nan=True) and \
    np.allclose(p2_old,p2,rtol=0.01,atol=0,equal_nan=True) and \
    np.allclose(p3_old,p3,rtol=0.01,atol=0,equal_nan=True) and \
    np.allclose(p4_old,p4,rtol=0.01,atol=0,equal_nan=True) and \
    np.allclose(p5_old,p5,rtol=0.01,atol=0,equal_nan=True):
        flag="false"
    else:
        p1_old,p2_old,p3_old,p4_old,p5_old=p1,p2,p3,p4,p5
        epoch+=1
        print("\nepoch ",epoch,":\n")
        for k,row in enumerate(image1_reshaped):
            diff_p1=abs(np.sum(row-p1))
            diff_p2=abs(np.sum(row-p2))
            diff_p3=abs(np.sum(row-p3))
            diff_p4=abs(np.sum(row-p4))
            diff_p5=abs(np.sum(row-p5))
            
            if min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p1:
                image1_with_clusters[k]=1
                p1_cluster=np.vstack((p1_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p2:
                image1_with_clusters[k]=2
                p2_cluster=np.vstack((p2_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p3:
                image1_with_clusters[k]=3
                p3_cluster=np.vstack((p3_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p4:
                image1_with_clusters[k]=4
                p4_cluster=np.vstack((p4_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p5:
                image1_with_clusters[k]=5
                p5_cluster=np.vstack((p5_cluster,row))
        p1,p2,p3,p4,p5=np.mean(p1_cluster, axis=0),np.mean(p2_cluster, axis=0),np.mean(p3_cluster, axis=0),np.mean(p4_cluster, axis=0),np.mean(p5_cluster, axis=0)
        print(p1)
        print(p2)
        print(p3)
        print(p4)
        print(p5)
```

```python
for l in range(image1_reshaped.shape[0]):
    if image1_with_clusters[l]==1.0:
        image1_k[l]=p1
    elif image1_with_clusters[l]==2.0:
        image1_k[l]=p2
    elif image1_with_clusters[l]==3.0:
        image1_k[l]=p3
    elif image1_with_clusters[l]==4.0:
        image1_k[l]=p4
    elif image1_with_clusters[l]==5.0:
        image1_k[l]=p5
```

```python
image1_k=image1_k.reshape(image1.shape)
plt.imshow(image1_k)
plt.show()
```

### Analysis of Image 1

Through the application of k-mean clustering, it could be observed that for k=2, the image is being prominently divided into 2 groups; thus the regenerated image is depicted into two colors. Further incrementing the value of k, results in formation of more clusters; therefore, the regenerated image has better formation in terms of shape and color perception. Besides, through running the algorithm on various images, it is noted that the shading and quality of of image impact the results of k-mean approach applied. Good image quality with lesser fluctuations in the color shades gave out better regenerated images.
