## For k=5

```python
p1=np.random.randint(0,255,(1,3))
p2=np.random.randint(0,255,(1,3))
p3=np.random.randint(0,255,(1,3))
p4=np.random.randint(0,255,(1,3))
p5=np.random.randint(0,255,(1,3))
p1_cluster=p2_cluster=p3_cluster=p4_cluster=p5_cluster=np.empty((0,3))

epoch=0
flag="true"
print(p1)
print(p2)
print(p3)
print(p4)
print(p5)

while flag=="true":
    if np.allclose(p1_old,p1,rtol=0.0001,atol=0,equal_nan=True) and \
    np.allclose(p2_old,p2,rtol=0.0001,atol=0,equal_nan=True) and \
    np.allclose(p3_old,p3,rtol=0.0001,atol=0,equal_nan=True) and \
    np.allclose(p4_old,p4,rtol=0.0001,atol=0,equal_nan=True) and \
    np.allclose(p5_old,p5,rtol=0.0001,atol=0,equal_nan=True):
        flag="false"
    else:
        p1_old,p2_old,p3_old,p4_old,p5_old=p1,p2,p3,p4,p5
        epoch+=1
        print("\nepoch ",epoch,":\n")
        for row in image1:
            diff_p1=abs(np.sum(row-p1))
            diff_p2=abs(np.sum(row-p2))
            diff_p3=abs(np.sum(row-p3))
            diff_p4=abs(np.sum(row-p4))
            diff_p5=abs(np.sum(row-p5))
            
            if min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p1:
                p1_cluster=np.vstack((p1_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p2:
                p2_cluster=np.vstack((p2_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p3:
                p3_cluster=np.vstack((p3_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p4:
                p4_cluster=np.vstack((p4_cluster,row))
            elif min(diff_p1,diff_p2,diff_p3,diff_p4,diff_p5)==diff_p5:
                p5_cluster=np.vstack((p5_cluster,row))
                
        p1,p2,p3,p4,p5=np.mean(p1_cluster, axis=0),np.mean(p2_cluster, axis=0),np.mean(p3_cluster, axis=0),np.mean(p4_cluster, axis=0),np.mean(p5_cluster, axis=0)
        print(p1)
        print(p2)
        print(p3)
        print(p4)
        print(p5)
```

```python
plt.scatter(p1_cluster[: , 0], p1_cluster[: , 1], s=100, c='red', label ='Cluster 1')
plt.scatter(p2_cluster[: , 0], p2_cluster[: , 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(p3_cluster[: , 0], p3_cluster[: , 1], s=100, c='green', label ='Cluster 3')
plt.scatter(p4_cluster[: , 0], p4_cluster[: , 1], s=100, c='cyan', label ='Cluster 4')
plt.scatter(p5_cluster[: , 0], p5_cluster[: , 1], s=100, c='magenta', label ='Cluster 5')
```
