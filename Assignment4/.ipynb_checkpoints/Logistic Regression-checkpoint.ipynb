### Test Data Loading

```python
test_input = np.loadtxt('testX.txt')
test_output = np.loadtxt('testY.txt')
```

```python
train_input_ones = np.ones(train_input.shape[0])
test_input_ones = np.ones(test_input.shape[0])
train_input = np.insert(train_input, 0, train_input_ones, axis=1)
test_input = np.insert(test_input, 0, test_input_ones, axis=1)
```

```python
train_output = np.where(train_output==4,0,train_output)
train_output = np.where(train_output==2,1,train_output)
test_output = np.where(test_output==4,0,test_output)
test_output = np.where(test_output==2,1,test_output)

print(train_output)
print(test_output)
```

```python
def logistic_regression(train_input,train_output, alpha):
    train_weights = np.random.uniform(low=-1, high=1, size=(train_input.shape[1],))
    for i in range(1000):
        train_prediction = 1/(1+np.exp(-1*np.dot(train_input, train_weights)))         
        train_cost = (-1/train_input.shape[0]) * (np.sum((train_output.T * np.log(train_prediction)) + ((1-train_output.T) * np.log(1-train_prediction))))
        train_weights = train_weights - (alpha/train_input.shape[0]) * np.dot(train_input.T,(train_prediction - train_output))
        
        #-------------------------------------------------------------------------
        #print(train_prediction)
        updated_train_prediction = np.where(train_prediction>0.5,1,train_output)
        updated_train_prediction = np.where(train_prediction<0.5,0,train_output)
        #print("new:",train_prediction)
      
        train_correct=0
        correct_2=0;
        correct_4=0;
        for n,value in enumerate(updated_train_prediction):
            if value==1 and train_output[n]==1:
                train_correct=train_correct+1
                correct_4+=1
            elif value==0 and train_output[n]==0:
                train_correct=train_correct+1
                correct_2+=1
                
        train_accuracy=(train_correct)/train_input.shape[0]

        #--------------------------------------------------------------
        print('Epoch: {} \tAccuracy: {:.6f}'.format(i, train_accuracy*100))
        
        #-----------------------------------------------------------------
        
        
        
    return train_weights
```

```python
w= logistic_regression(train_input,train_output, 0.01)
```
